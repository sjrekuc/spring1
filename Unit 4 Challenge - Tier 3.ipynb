**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# create a DF for 1998
df98 = dfa.loc[dfa['Year']==1998]
## df_98_18 = dfa.loc[dfa['Year']==1998]
# create another DF for 2018
df18 = dfa.loc[dfa['Year']==2018]
# also add to the df
## df_98_18['price_2018'] = dfa.loc[dfa['Year']==2018].Average_price
## try calculating the price ratio for one burough (yes, I realize it would have been easier converting to Numpy)
df18.loc[df18['Borough']=='Barking & Dagenham'].Average_price.values[0] / df98.loc[df98['Borough']=='Barking & Dagenham'].Average_price.values[0]
```

These seems about right. Now, I need to do it for each Borough.

```python
# use a dictionary to capture these values
ratio = {}
# loop through the borough names and find the ratio of 2018 / 1998
for b in df18['Borough']:
    r = df18.loc[df18['Borough']==b].Average_price.values[0] / df98.loc[df98['Borough']==b].Average_price.values[0]
    # add to dictionary as a 1 item list to make the DF easier
    ratio[b] = [r]
# print the ratios to make sure they make sense
print(ratio)
```

```python
## create a DF of the ratio and print out the DF to make sure it looks good
dfr = pd.DataFrame(ratio)
dfr.head()
```

```python
## Transpose the DF and reset index; rename columns appropriately
dfr_t = dfr.transpose()
dfr_t = dfr_t.reset_index()
dfr_t.rename(columns={'index':'Borough', 0:'Ratio'}, inplace=True)
dfr_t.head()
```

```python
dfrts = dfr_t.sort_values(by='Ratio',ascending=False)
dfrts
```

```python
## Plot DF 
dfrts.plot(x='Borough', y='Ratio', kind='bar', figsize=(20,10))
plt.show()
```
